```python
import time, requests, json, pandas as pd
```

```python
api_key = '49116535-0a13-4c93-aaf9-d3049472f3e8'
headers = {'Authorization': api_key, 'Content-type': 'application/json'}
api_addr = 'http://ec2-3-113-32-69.ap-northeast-1.compute.amazonaws.com/api/action/'

def create_dataset_package(package_metadata):
    url = 'http://localhost/api/action/package_create'
    package_data = package_metadata
    
    # Create dataset package
    package = requests.post(url, headers=headers, data=json.dumps(package_data))

    # Check for created package id; 6ede2d45-193d-443d-8ec6-3ff8c3b13d9a
    return json.loads(package.content)

def add_resource(path_to_resource, resource_metadata):
    headers = {'Authorization': api_key}
    resource_data = resource_metadata
    url = 'http://localhost/api/action/resource_create'
    
    resource = requests.post(url, data=resource_data, headers=headers, \
                            files=[('upload', open(path_to_resource, encoding='utf-8'))])
    
    return json.loads(resource.content)

def delete_resource(resource_id):
    url = 'http://localhost/api/action/resource_delete'
    
    delete = requests.post(url, headers=headers, data=json.dumps(resource_id))
    
    return json.loads(delete.content)

def delete_dataset_package(dataset_package_id):
    url = 'http://localhost/api/action/package_delete'
    
    delete = requests.post(url, headers=headers, data=json.dumps(dataset_package_id))
    
    return json.loads(delete.content)
    
def delete_table_row(resource_id, filters=None):
    url = 'http://localhost/api/action/datastore_delete'
    
    if filters:
        delete_data = {'resource_id': resource_id,
                       'force': True,
                       'filters': filters}
    else:
        delete_data = {'resource_id': resource_id,
                       'force': True}
        
    delete = requests.post(url, headers=headers, data=json.dumps(delete_data))
    
    return json.loads(delete.content)

# def get_current_info(organization=False, packages=False, groups=False, users=False):
#     results = []
#     org = 'organization_list'
#     pack = 'package_list'
#     grp = 'group_list'
#     users = 'user_list'
    
#     if organization:
#         return json.loads(requests.get(api_addr+org).content)
    
        
    
# def dataset_table(path_to_dataset, dataset_metadata):
def dataset_table(path_to_dataset):
    url = 'http://localhost/api/action/datastore_create'
    records = []
    data = pd.read_csv(path_to_dataset)

    fields = [{'id': 'index', 'type': 'bigint'},
              {'id': 'country', 'type': 'text'},
              {'id': 'id.val', 'type': 'text'},
              {'id': 'id.type', 'type': 'text'},
              {'id': 'events', 'type': 'text'},
              {'id': 'region', 'type': 'text'}
             ]

    for i in range(len(data)):
        record = {'index': int(data['index'][i]),
                  'country': data['country'][i],
                  'id.val': data['id.val'][i],
                  'id.type': data['id.type'][i],
                  'events': data['events'][i],
                  'region': data['region'][i]
        }
        records.append(record)
        
    table_data = {'resource': '79db2c33-cb75-4ab9-8ebe-70e4734984f7',
                  'force': True,  
                  'fields': fields, 
                  'records': records, 
                  'primary_key': ['index']
                 }
        
    # Create table for dataset
    data_table = requests.post(url, headers=headers, data=json.dumps(table_data))
    
    return json.loads(data_table.content)
```
