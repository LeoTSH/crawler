### Import packages and set data path

```python
import time, json, pandas as pd, numpy as np
from nhattao_crawler import Nhattao_crawler
from datetime import datetime

data_path = '../data/nhattao/'
```

### Crawl Marketplace

```python
nhattao_crawler = Nhattao_crawler('https://nhattao.com/f/can-mua-trao-doi.589/')
no_pages = nhattao_crawler.get_no_pages()
combined = []
start = time.time()

for i in range(1, no_pages+1):
    print('Processing page {} of {}'.format(i, no_pages))
    
    print('Retrieving links')
    page_listings = nhattao_crawler.get_listings_per_page(i)
    for j in range(len(page_listings)):
        sleep_timer = np.random.randint(8, 16)
        print('Processing page {} of {}, link {} of {}'.format(i, no_pages, j+1, len(page_listings)))
        print(page_listings[j])    
        
        print('Retrieving link data')
        combined.append(nhattao_crawler.process_data(page_listings[j]))
        
        # Sleep to prevent timeout
        print('Sleeping for {} seconds'.format(sleep_timer))
        print('\n')
        time.sleep(sleep_timer)
print('Scrape completed')
print('Total time taken: {:.2f}'.format(time.time()-start))
```

### Serialize marketplace data to json

```python
print('Writing json file')
with open(data_path+str(datetime.timestamp(datetime.now()))+'_nhattao_marketplace.json', 'w') as f:
    json.dump(combined, f)
```

### Batdongsan

```python
import time, json, pandas as pd, numpy as np, requests, re
from bs4 import BeautifulSoup
from datetime import datetime

data_path = '../data/batdongsan/'
```

```python
def process_details(soup):
    k, v = [], []
    results = {}
    
    soup = soup.find('div', class_='div-table')

    for key in soup.find_all('div', class_='left'):
        k.append(key.text.strip())
    for value in soup.find_all('div', class_='right'):
        v.append(value.text.replace('\r\n','').strip())
        
    for i in range(len(k)):
        if k[i] == 'Email':
            continue
        else:
            results[k[i]] = v[i]
    return results
    
def get_posted_date(soup):
    dt = soup.find('div', class_='prd-more-info').text
    dt = re.findall(r'\d{2}-\d{2}-\d{4}', dt)
    dt = datetime.timestamp(datetime.strptime(dt[-2], '%d-%m-%Y'))
    return dt
```

```python
combined = []
no_pages = 50
start = time.time()
url = 'https://batdongsan.com.vn/nha-dat-ban/'
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}
s = requests.Session()


for i in range(1, 3): # Change here
    print('Processing page {} of {}'.format(i, no_pages))
    
    print('Retrieving links')
    r = s.get(url+'p'+str(i), headers=headers)
    soup = BeautifulSoup(r.text, 'html.parser')
    links = []
    for x in soup.find_all('div', class_='vip0 search-productItem'):
        links.append('https://batdongsan.com.vn'+x.a['href'])
        
    for j in range(len(links)):
        sleep_timer = np.random.randint(8, 16)
        print('Processing page {} of {}, link {} of {}'.format(i, no_pages, j+1, len(links)))
        print(links[j])
        
        r = s.get(links[j], headers=headers)
        soup = BeautifulSoup(r.text, 'html.parser')
        price_area = soup.find_all('span', class_='gia-title')
        results = {'Listing URL': links[j], 
                   'Listing ID': re.findall(r'\d', links[j].split('-')[-1]),
                   'Title': soup.find('div', class_='pm-title').text.strip(),
                   'District': soup.find('span', class_='diadiem-title mar-right-15').text.split(':')[-1].strip(),
                   'Price': price_area[0].strong.text.strip(), 
                   'Area': price_area[1].strong.text.strip(),
                   'Posted Date': get_posted_date(soup)
                  }
        details = process_details(soup)        
        for k, v in details.items():
                results[k] = v
        combined.append(results)
        time.sleep(sleep_timer)
print('Scrape completed')
print('Total time taken: {:.2f}'.format(time.time()-start))
```

### Serialize Batdongsan data

```python
print('Writing json file')
with open(data_path+str(datetime.timestamp(datetime.now()))+'_batdongsan_json', 'w') as f:
    json.dump(combined, f)
```
