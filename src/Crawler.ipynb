### Nhattao

### Import packages and set data path

```python
import time, json, pandas as pd, numpy as np
from nhattao_crawler import Nhattao_crawler
from datetime import datetime

data_path = '../data/nhattao/'
```

### Crawl Marketplace

```python
crawler = Nhattao_crawler('https://nhattao.com/f/can-mua-trao-doi.589/')
no_pages = crawler.get_no_pages()
combined = []
start = time.time()

for i in range(1, no_pages+1):
    print('Processing page {} of {}'.format(i, no_pages))
       
    page_listings = crawler.get_listings_per_page(i)
    for j in range(len(page_listings)):
        sleep_timer = np.random.randint(5, 11)
        print('Processing page {} of {}, link {} of {}'.format(i, no_pages, \
                                                               j+1, len(page_listings)))
        print(page_listings[j])
        print('\n')
        
        crawler.process_data(page_listings[j])
        combined.append(crawler.process_data(page_listings[j]))
        
        # Sleep to prevent timeout
        print('Sleeping for {} seconds'.format(sleep_timer))
        time.sleep(sleep_timer)
print('Scrape completed')
print('Total time taken: {:.2f}'.format(time.time()-start))
```

### Serialize marketplace data to json

```python
print('Writing json file')
with open(data_path+str(datetime.timestamp(datetime.now()))+'_nhattao_marketplace.json', 'w') as f:
    json.dump(combined, f)
```

```python
test_pd = pd.read_json(data_path+'1569841832.446193_nhattao_marketplace.json')
test_pd.to_excel(data_path+str(datetime.timestamp(datetime.now()))+'_nhattao_marketplace.xlsx', \
                 index=False)
```

### Chotot

```python
import time, requests, time, numpy as np, pandas as pd, re
from datetime import datetime, timedelta
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
```

```python
chrome_opts = Options()
chrome_opts.headless = True
```

```python
def process_details(soup):
    process_details = {}
    for details in soup.find_all('div', class_='col-xs-12 col-md-6 _2E8caC-j61im7lRi5lExI9'):
        if len(details.text.split(':')) > 1:
            details = details.text.split(':')        
            process_details[details[0].strip()] = details[1].strip()
        else:
            process_details[details.text] = details.text
    return process_details

def process_datetime(soup):
    dt = soup.find('div', class_='hidden-xs JfuoT2phEEouoxezYbBx4').text.split()[-3:-1]
    
```

```python
# Set main url page
url = 'https://nha.chotot.com/#regionselect?v=1.1'
no_pages = 50

# Set driver for selenium
driver = webdriver.Chrome('./win_chromedriver.exe', options=chrome_opts)

# Retrieve webpage
driver.get(url)

# Parse retrieved page 
page_links = BeautifulSoup(driver.page_source, 'html.parser')

# Retrieve all page links
pages = set()
for page in page_links.find_all('li', class_='sc-dVhcbM cZLfdN'):
    pages.add(page.a.get('href'))
pages = list(pages)

# Cycle through each page link
combined = []
for i in range(0, len(pages)+1): # Change here 
    links = []
    page_url = 'https://nha.chotot.com'+pages[i] 
    print('Processing page {} of {}'.format(i+1, len(pages)))
    print(page_url)
       
    driver.get(page_url)    
    post_links = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Retrieve all posting links from the page
    for link in post_links.find_all('a', href=re.compile(r'.*.htm')):
        links.append('https://nha.chotot.com'+link.get('href'))
    
    for j in range(len(links)):
        print('Processing link {} out of {}'.format(j+1, len(links)))
        print('Link: {}'.format(links[j]))
        driver.get(links[j])
        # Sleep to allow webpage to load
        time.sleep(5)
        posting_info = BeautifulSoup(driver.page_source, 'html.parser')
        results = {
            'Link': links[j],
            'Thread ID': links[j].split('/')[-1].replace('.htm',''),
            'Seller': posting_info.find('div', class_='sc-eilVRo hEXVti').text,
            'Seller Type': posting_info.find('div', class_='sc-fhYwyz fDMLIV').text,
            'Contact': posting_info.find('a', \
                        class_='H3QBvet3qzdHlB3LVAw-7 btn btn-success hidden-sm hidden-md hidden-lg hidden-xl')\
                        ['href'].split(':')[1],
            'Title': posting_info.find('h1', class_='_22kG1zbJ4D-6IUEgKvoifC col-xs-12').text,
            'Price': posting_info.find('span', \
                    class_='oRSYZ0HPb2tHhHjpVp_2o').text.replace(' VND', '').replace('', '').replace('.','')
                 }
        details = process_details(posting_info)
        print('\n')
        for k, v in details.items():
            results[k] = v
        combined.append(results)
        sleep_timer = np.random.randint(5, 11)
        print('Sleeping for: {} seconds'.format(sleep_timer))
        time.sleep(sleep_timer)
driver.quit()
```

```python
print('Writing json file')
with open('../data/chotot/'+str(datetime.timestamp(datetime.now()))+'_chotot_test.json', 'w') as f:
    json.dump(combined, f)
```
